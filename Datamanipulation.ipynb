```python
# for data visualization
import pandas as pd
import matplotlib.pyplot as plt
mydata={'Date':['01/01/24','01/02/24','01/03/24'],
        'Price':[100,120,150]}
print(mydata)
df=pd.DataFrame(mydata)
print(df)
df['Date']=pd.to_datetime(df['Date'])
df.plot(x= 'Date' , y='Price', kind = 'line')
plt.xlabel='Date'
plt.ylabel='Price'
plt.show()
```

```python
#establish relation between two number variables using scatter plot
#import pandas library
import pandas as pd
import matplotlib.pyplot as plt
#define and declare dictionary
mydata={'Study hours':[2,4,1,7],'Marks':[20,60,10,80]}
#print(mydata)
df=pd.DataFrame(mydata)
#print(df)
df.plot(x ='Study hours',y ='Marks',kind ='scatter')
plt.show()
```

```python
#How to save the DF into a file
import pandas as pd
mydata={'Name':['Rahul','Anish'],
        'Rank':[1,2]}
#print(mydata)
df=pd.DataFrame(mydata)
print(df)
#df.to_csv('mydata')
#df.to_json('mydatajson')
df.to_parquet('newparfile')
```

```python
#find the sum of a column from a DF file with orderID and price
mydata={'orderid':[101,201,301,401,501],
        'price':[1000,2000,3000,4000,5000]}
#print(mydata)
import pandas as pd
df=pd.DataFrame(mydata)
print(df)
df.aggregate({'price':['sum']})
```
