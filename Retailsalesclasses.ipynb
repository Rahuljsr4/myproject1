```python
#This program will get the csv file and perform analysis on data with presentation using  Classes and objects
class retailsalesanalyzer:  #defining the class
  # DATA LOAD - load the csv file
  def __init__(prd,csvfile):
    prd.df=pd.read_csv(csvfile)
  #  print(prd.df)
  # DATA CLEANING - Change NaN values to null
  def clean(prd): #always pass reference to the attributes which is data frame for the csv file data.
    prd.df.fillna(0,inplace=True)
    return (print(prd.df))
  # DATA MANIPULATION
  #calculate sales per product
  def salesbyproduct(prd): #passing the reference to the object
    salebyprod=prd.df.groupby('Product')['Sales'].sum()    #return salebyprod
    return (print(salebyprod))
  #calculate avg sales per product
  def avg_daily_sale(prd):
    avgsale=prd.df.groupby('Product')['Sales'].mean()
    return(avgsale)
  #VISUALIZATION
  #plot sales trend over the time
  def salestrend(prd):
    salestrend=prd.df.groupby('Date')['Sales'].sum()
    salestrend.plot(x='Date',y='Sales',kind='line')
    return(plt.show())
  #plot sales per product using bar graph
  def salesperprod(prd):
    salesprod= prd.df.groupby('Product')['Sales'].sum()
    salesprod.plot(x='Product',y='Sales',kind='bar')
    return(plt.show())

#Main body
import pandas as pd
import matplotlib.pyplot as plt
callclass=retailsalesanalyzer('/content/retail_sales copy.csv') #declaring the object for the class
callclass.clean()
callclass.salesbyproduct()
callclass.avg_daily_sale()
callclass.salestrend()
callclass.salesperprod()
```
